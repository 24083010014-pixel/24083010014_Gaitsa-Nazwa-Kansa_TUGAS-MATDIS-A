###**Abstract**

This research investigates the influence of product quality, distinctiveness, and consumer satisfaction on consumer loyalty, as well as the mediating role of consumer satisfaction in the relationship between product quality, distinctiveness, and loyalty. The study was conducted with 190 customers of Sambel Pecel Karangsari Blitar, a traditional chili paste producer, using a combination of accidental and purposive sampling to ensure respondents matched the research criteria. Data were analyzed using Ordinary Least Squares (OLS) and path analysis to determine both direct and indirect effects. The findings reveal that product quality, distinctiveness, and consumer satisfaction all have significant positive effects on consumer loyalty. Direct effects of product quality and distinctiveness on loyalty are stronger than their indirect effects through consumer satisfaction, which is confirmed as a partial mediator by the Sobel test. The study emphasizes that maintaining consistent product quality, unique and distinctive characteristics, and high consumer satisfaction is crucial for fostering long-term loyalty. These insights are particularly relevant for traditional food businesses, where trust, authenticity, and brand identity play a key role in retaining customers in a competitive market.

###**Introduction**

In today’s competitive business environment, maintaining long-term consumer loyalty is one of the most challenging yet crucial aspects of sustaining a successful business. The existence and continuity of a business depend largely on consumers who repeatedly purchase and support its products. As noted in consumer behavior theory, loyalty is built upon satisfaction, perceived benefits, and a willingness to repurchase in the future. When consumers experience consistent product performance that meets their expectations, they are more likely to remain loyal.

From the consumer’s perspective, product quality is not defined by the producer but by how consumers evaluate the product’s ability to meet their expectations. A product that consistently maintains its expected quality will earn trust and encourage repeat purchases. Likewise, distinctiveness—the uniqueness or special characteristics that differentiate a product—serves as a psychological and emotional trigger that attracts consumers. According to Kotler and Keller (2012), distinctive features allow a brand to stand out in consumers’ memory, making it more recognizable and preferred over competitors.

This research focuses on Sambel Pecel Karangsari, a traditional chili paste business located in Blitar, Indonesia, that has successfully built a strong brand identity over time. Karangsari is not only a well-known local brand but also the only Sambel Pecel producer with a registered trademark and export distribution channels. The company’s transformation from a home-based business into PT Karangsari Food Industri represents a rare achievement among small and medium-sized enterprises in the region.

The study aims to explore how product quality and distinctiveness contribute to consumer loyalty, and how consumer satisfaction mediates these effects. Understanding this relationship is essential, particularly for traditional food producers, because the market is saturated with similar products at low prices. Maintaining consumer trust and loyalty requires consistent quality and a unique identity that stands out from the competition

###**Conclusion**

The findings of this study demonstrate that product quality, distinctiveness, and consumer satisfaction significantly influence consumer loyalty. High-quality products enhance consumer trust and positive experiences, while unique and distinctive characteristics strengthen emotional connections with the brand. Together, these factors encourage repeat purchases and long-term loyalty.

Consumer satisfaction is found to play a partial mediating role between product quality, distinctiveness, and loyalty. This means that although satisfaction reinforces loyalty, it does not entirely account for it—other elements such as brand image, nostalgia, and inherited preferences also contribute to why consumers continue to choose the same product. The direct effects of product quality and distinctiveness on loyalty are stronger than the indirect effects through satisfaction, suggesting that the Karangsari brand has already established strong consumer confidence and familiarity independent of satisfaction alone.

Moreover, the study provides valuable insight into the behavior of consumers in local food industries. It shows that traditional food products like Sambel Pecel Karangsari can maintain loyal customers through consistent quality, authenticity, and brand identity. In essence, long-term loyalty is not just the result of satisfaction, but also of trust, tradition, and emotional attachment formed between consumers and the brand.

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns

abstract = """
This research investigates the influence of product quality, distinctiveness,
and consumer satisfaction on consumer loyalty, as well as the mediating role
of consumer satisfaction in the relationship between product quality,
distinctiveness, and loyalty. The study was conducted with 190 customers of
Sambel Pecel Karangsari Blitar, a traditional chili paste producer, using a
combination of accidental and purposive sampling to ensure respondents
matched the research criteria. Data were analyzed using Ordinary Least
Squares (OLS) and path analysis to determine both direct and indirect effects.
The findings reveal that product quality, distinctiveness, and consumer
satisfaction all have significant positive effects on consumer loyalty. Direct
effects of product quality and distinctiveness on loyalty are stronger than
their indirect effects through consumer satisfaction, which is confirmed as a
partial mediator by the Sobel test. The study emphasizes that maintaining
consistent product quality, unique and distinctive characteristics, and high
consumer satisfaction is crucial for fostering long-term loyalty. These
insights are particularly relevant for traditional food businesses, where trust,
authenticity, and brand identity play a key role in retaining customers in a
competitive market.
"""

introduction = """
In today’s competitive business environment, maintaining long-term consumer loyalty
is one of the most challenging yet crucial aspects of sustaining a successful business.
The existence and continuity of a business depend largely on consumers who repeatedly
purchase and support its products. As noted in consumer behavior theory, loyalty is built
upon satisfaction, perceived benefits, and a willingness to repurchase in the future. When
consumers experience consistent product performance that meets their expectations, they
are more likely to remain loyal. From the consumer’s perspective, product quality is not
defined by the producer but by how consumers evaluate the product’s ability to meet their
expectations. A product that consistently maintains its expected quality will earn trust and
encourage repeat purchases. Likewise, distinctiveness—the uniqueness or special
characteristics that differentiate a product—serves as a psychological and emotional trigger
that attracts consumers. According to Kotler and Keller (2012), distinctive features allow
a brand to stand out in consumers’ memory, making it more recognizable and preferred
over competitors. This research focuses on Sambel Pecel Karangsari, a traditional chili paste
business located in Blitar, Indonesia, that has successfully built a strong brand identity over
time. Karangsari is not only a well-known local brand but also the only Sambel Pecel
producer with a registered trademark and export distribution channels. The company’s
transformation from a home-based business into PT Karangsari Food Industri represents
a rare achievement among small and medium-sized enterprises in the region. The study
aims to explore how product quality and distinctiveness contribute to consumer loyalty,
and how consumer satisfaction mediates these effects. Understanding this relationship is
essential, particularly for traditional food producers, because the market is saturated with
similar products at low prices. Maintaining consumer trust and loyalty requires consistent
quality and a unique identity that stands out from the competition.
"""

conclusion = """
The findings of this study demonstrate that product quality, distinctiveness, and consumer
satisfaction significantly influence consumer loyalty. High-quality products enhance consumer
trust and positive experiences, while unique and distinctive characteristics strengthen
emotional connections with the brand. Together, these factors encourage repeat purchases
and long-term loyalty. Consumer satisfaction is found to play a partial mediating role between
product quality, distinctiveness, and loyalty. This means that although satisfaction reinforces
loyalty, it does not entirely account for it—other elements such as brand image, nostalgia,
and inherited preferences also contribute to why consumers continue to choose the same
product. The direct effects of product quality and distinctiveness on loyalty are stronger than
the indirect effects through satisfaction, suggesting that the Karangsari brand has already
established strong consumer confidence and familiarity independent of satisfaction alone.
Moreover, the study provides valuable insight into the behavior of consumers in local food
industries. It shows that traditional food products like Sambel Pecel Karangsari can maintain
loyal customers through consistent quality, authenticity, and brand identity. In essence,
long-term loyalty is not just the result of satisfaction, but also of trust, tradition, and emotional
attachment formed between consumers and the brand.
"""

text = abstract + introduction + conclusion

# Ambil semua kata (huruf saja)
words = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Fungsi hitung vokal & konsonan
def count_vowels(word):
    return sum(1 for c in word if c in 'aeiou')

def count_consonants(word):
    return sum(1 for c in word if c.isalpha() and c not in 'aeiou')

# Buat DataFrame
data = pd.DataFrame({
    'word': words,
    'vowels': [count_vowels(w) for w in words],
    'consonants': [count_consonants(w) for w in words]
})

# Pivot tabel frekuensi total kata
pivot_freq = data.pivot_table(
    index='vowels', columns='consonants',
    aggfunc='size', fill_value=0
)

# Pivot tabel jumlah kata unik
pivot_unique = data.groupby(['vowels', 'consonants'])['word'].nunique().unstack(fill_value=0)

# Tampilkan tabel
print("Pivot Tabel: Frekuensi Kombinasi Konsonan-Vokal")
print(pivot_freq)

print("\nPivot Tabel: Jumlah Kata Unik")
print(pivot_unique)

# Heatmap visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.heatmap(pivot_freq, cmap='YlGnBu')
plt.title('Frekuensi Kombinasi Konsonan–Vokal')
plt.show()

plt.figure(figsize=(8,5))
sns.heatmap(pivot_unique, cmap='magma')
plt.title('Jumlah Kata Unik per Kombinasi Konsonan–Vokal')
plt.show()
```

# 2. Jumlah Kata dalam Kalimat

```python
import pandas as pd
import re
import matplotlib.pyplot as plt

# Fungsi menghitung jumlah kata per kalimat
def sentence_word_counts(text):
    sentences = [s.strip() for s in re.split(r'[.!?]', text) if s.strip()]
    return [len(re.findall(r'\b[a-zA-Z]+\b', s)) for s in sentences]

# Hitung jumlah kata tiap kalimat untuk tiap bagian
data = pd.DataFrame({
    "Bagian": ["Abstract"] * len(sentence_word_counts(abstract))
              + ["Introduction"] * len(sentence_word_counts(introduction))
              + ["Conclusion"] * len(sentence_word_counts(conclusion)),
    "Jumlah Kata": sentence_word_counts(abstract)
                  + sentence_word_counts(introduction)
                  + sentence_word_counts(conclusion)
})

# Tampilkan tabel rapi
print("📊 Jumlah Kata per Kalimat di Tiap Bagian:")
print(data)

# Histogram per bagian
plt.figure(figsize=(8,5))
for section in ["Abstract", "Introduction", "Conclusion"]:
    subset = data[data["Bagian"] == section]
    plt.hist(subset["Jumlah Kata"], bins=range(0, max(data["Jumlah Kata"])+5, 5),
             alpha=0.6, label=section)

plt.title("Distribusi Jumlah Kata per Kalimat di Tiap Bagian")
plt.xlabel("Jumlah Kata per Kalimat")
plt.ylabel("Frekuensi")
plt.legend()
plt.show()
```

# 3. Jumlah Kalimat per Paragraf

```python
import matplotlib.pyplot as plt
import re

# Teks yang sudah dipisah
sections = {
    "Abstract": abstract,
    "Introduction": introduction,
    "Conclusion": conclusion
}

sentence_counts = {}

# Hitung jumlah kalimat tiap paragraf
for section, text in sections.items():
    # Pisahkan paragraf berdasarkan newline kosong
    paragraphs = [p.strip() for p in text.split('\n\n') if p.strip()]
    counts = []
    for para in paragraphs:
        # Hitung kalimat berdasarkan titik (.), tanda tanya (?), atau tanda seru (!)
        sentences = re.split(r'[.!?]', para)
        # Hapus elemen kosong
        sentences = [s for s in sentences if s.strip()]
        counts.append(len(sentences))
    sentence_counts[section] = counts

# Cetak hasil per paragraf
for section, counts in sentence_counts.items():
    print(f"\n{section}:")
    for i, count in enumerate(counts, 1):
        print(f"  Paragraf {i}: {count} kalimat")

# Buat histogram
plt.figure(figsize=(12,6))
colors = ['skyblue', 'lightgreen', 'salmon']

for i, (section, counts) in enumerate(sentence_counts.items()):
    plt.bar([x + i*0.25 for x in range(len(counts))], counts, width=0.25, color=colors[i], label=section)

plt.xlabel('Paragraf')
plt.ylabel('Jumlah Kalimat')
plt.title('Jumlah Kalimat per Paragraf di Setiap Bagian')
plt.xticks([x + 0.25 for x in range(max(len(c) for c in sentence_counts.values()))],
           [f"P{x+1}" for x in range(max(len(c) for c in sentence_counts.values()))])
plt.legend()
plt.show()
```

# Pairplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

data = {'Bagian': ['Abstract', 'Introduction', 'Conclusion'],
        'text': [abstract, introduction, conclusion]}
df = pd.DataFrame(data)

# Fungsi analisis per kalimat
def analyze_sentences(text):
    sentences = re.split(r'[.!?]', text)
    data = []
    for s in sentences:
        s = s.strip()
        if len(s) > 0:
            words = s.split()
            num_chars = len(s.replace(" ", ""))
            num_vowels = len(re.findall(r'[aeiouAEIOU]', s))
            num_consonants = num_chars - num_vowels
            data.append({
                'Kalimat': s,
                'Panjang_Kalimat': len(words),
                'Rata2_Panjang_Kata': num_chars / len(words),
                'Vokal': num_vowels,
                'Konsonan': num_consonants
            })
    return pd.DataFrame(data)

# Pisahkan berdasarkan bagian dan analisis
dfs = []
for index, row in df.iterrows():
    temp = analyze_sentences(row['text'])
    temp['Bagian'] = row['Bagian']
    dfs.append(temp)

final_df = pd.concat(dfs, ignore_index=True)

# Pairplot
sns.pairplot(final_df[['Vokal','Konsonan','Rata2_Panjang_Kata','Panjang_Kalimat','Bagian']],
             hue='Bagian', diag_kind='kde', corner=True)
plt.suptitle("Hubungan antara Vokal, Konsonan, Panjang Kata, dan Panjang Kalimat", y=1.02)
plt.show()
```

Berdasarkan visualisasi pairplot, terlihat adanya pola hubungan yang kuat antarvariabel linguistik di ketiga bagian artikel (Abstract, Introduction, Conclusion).

Terdapat korelasi positif yang kuat antara jumlah vokal, konsonan, dan panjang kalimat; semakin panjang kalimat, semakin banyak huruf vokal dan konsonan. Rata-rata panjang kata juga cenderung meningkat seiring panjang kalimat, meski hubungannya tidak sekuat vokal-konsonan.

Berdasarkan bagian artikel:

* **Abstract (biru)**: sebaran titik paling luas, menunjukkan variasi panjang kalimat tinggi dan kompleks.
* **Introduction (oranye)**: distribusi terkonsentrasi, menunjukkan struktur kalimat yang konsisten.
* **Conclusion (hijau)**: variasi cukup besar, dengan beberapa kalimat pendek dan beberapa panjang.

Secara keseluruhan, panjang kalimat paling memengaruhi jumlah vokal dan konsonan, sementara rata-rata panjang kata memberi gambaran kompleksitas kalimat tiap bagian.
