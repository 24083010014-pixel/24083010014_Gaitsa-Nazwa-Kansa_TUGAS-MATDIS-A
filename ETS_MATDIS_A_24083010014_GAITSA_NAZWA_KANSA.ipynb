###**Abstract**

This research aims to analyze the effect of product quality, distinctiveness, and consumer satisfaction on consumer loyalty; examine the effect of product quality and distinctiveness on consumer loyalty mediated by consumer satisfaction. This research employed a quantitative approach by using 190 customers of Sambel Pecel Karangsari Blitar as the sample. The sampling technique used in this research is accidental sampling in collecting the respondents and purposive sampling in choosing respondents with the criteria matched to the research objectives. Ordinary Least Square (OLS) and Path analysis are the methods used to conduct the research analysis. The results showed product quality had a significant positive influence on consumer loyalty; distinctiveness had a significant positive influence on consumer loyalty; consumer satisfaction had a significant positive influence on consumer loyalty; product quality had direct influence on consumer loyalty; distinctiveness had direct influence on consumer loyalty; there are direct and indirect influences of product quality on consumer loyalty through consumer satisfaction since the direct influence is greater than the indirect influence based on path analysis result; there is also a similar result of direct and indirect influence of distinctiveness on consumer loyalty through consumer satisfaction since the direct influence is greater than the indirect influence based on path analysis result. Sobel test proved that consumer satisfaction successfully acted as a partial mediating variable.

### **3. Heatmap Visualisasi**

Visualisasi pertama (frekuensi kombinasi) menampilkan gradasi warna **biru tua** hingga **kuning muda**, di mana area paling gelap (biru tua) menunjukkan kombinasi huruf yang paling sering muncul. Terlihat area padat warna berada di kisaran **1â€“3 vokal dan 2â€“4 konsonan**, menandakan dominasi kata dengan panjang sedang dan struktur seimbang.
Sementara itu, heatmap kedua (kata unik) menampilkan warna **oranye dan ungu**, dengan puncak intensitas di kisaran **2â€“3 vokal dan 4â€“5 konsonan**. Hal ini menegaskan bahwa variasi kosakata paling tinggi muncul pada struktur huruf tersebut. Pola diagonal yang tampak dari kiri atas ke kanan bawah menunjukkan adanya **hubungan proporsional** antara banyaknya konsonan dan vokal, artinya semakin banyak konsonan yang digunakan, biasanya jumlah vokalnya juga bertambah secara seimbang.

```python
# Analisis tambahan: Kata paling sering muncul untuk setiap kombinasi vokalâ€“konsonan
most_common_word = (
    data.groupby(['vowels', 'consonants'])['word']
    .apply(lambda x: x.value_counts().index[0])
    .reset_index(name='most_common_word')
)

# Frekuensi kemunculan kata tersebut
most_common_count = (
    data.groupby(['vowels', 'consonants'])['word']
    .apply(lambda x: x.value_counts().iloc[0])
    .reset_index(name='count')
)

# Gabungkan keduanya
common_summary = pd.merge(most_common_word, most_common_count,
                          on=['vowels', 'consonants'])

print("\nTabel Kata Paling Umum per Kombinasi Vokalâ€“Konsonan:")
print(common_summary.sort_values(by='count', ascending=False).head(10))
```

Tabel Rasio Konsonan-Vokal per Kata

```python
# Definisi huruf vokal dan konsonan (case-insensitive)
vowels = set("aiueo")
consonants = set("bcdfghjklmnpqrstvwxyz")

# Bersihkan teks dari tanda baca dan ubah jadi huruf kecil
words = re.findall(r"[a-zA-Z]+", text.lower())

# Hitung jumlah vokal dan konsonan per kata
data = []
for word in words:
    v = sum(1 for c in word if c in vowels)
    k = sum(1 for c in word if c in consonants)
    ratio = round(k / v, 2) if v > 0 else None
    if ratio is not None:
        if ratio > 1.5:
            kategori = "Konsonan dominan"
        elif ratio < 0.75:
            kategori = "Vokal dominan"
        else:
            kategori = "Seimbang"
    else:
        kategori = "Tidak ada vokal"
    data.append([word, k, v, ratio, kategori])

# Buat dataframe
df = pd.DataFrame(data, columns=["Kata", "Konsonan", "Vokal", "Rasio (K/V)", "Kategori"])

# Tampilkan hanya 10 kata pertama
print("=== 10 Kata Pertama ===")
print(df.head(10).to_string(index=False))

# Buat ringkasan kategori
summary = df["Kategori"].value_counts().reset_index()
summary.columns = ["Kategori", "Jumlah Kata"]
print("\n=== Ringkasan Kategori ===")
print(summary.to_string(index=False))
```

Data tersebut menunjukkan bahwa dari 10 kata pertama dalam teks, sebagian besar memiliki rasio konsonan terhadap vokal yang lebih tinggi, sehingga dikategorikan sebagai **â€œKonsonan dominan.â€**
Secara keseluruhan, hasil ringkasan menunjukkan pola yang sama: kata **konsonan dominan (974 kata)** jauh lebih banyak dibandingkan kata **seimbang (774 kata)** dan **vokal dominan (77 kata)**.
Hal ini menandakan bahwa struktur kata dalam teks cenderung **padat dan kompleks**, khas tulisan akademik yang menggunakan banyak istilah teknis dan formal dengan jumlah konsonan lebih banyak daripada vokal.

# 2. Jumlah Kata dalam Kalimat

```python
import pandas as pd
import re
import matplotlib.pyplot as plt

# Fungsi menghitung jumlah kata per kalimat
def sentence_word_counts(text):
    sentences = [s.strip() for s in re.split(r'[.!?]', text) if s.strip()]
    return [len(re.findall(r'\b[a-zA-Z]+\b', s)) for s in sentences]

# Hitung jumlah kata tiap kalimat untuk tiap bagian
data = pd.DataFrame({
    "Bagian": ["Abstract"] * len(sentence_word_counts(abstract))
              + ["Introduction"] * len(sentence_word_counts(introduction))
              + ["Conclusion"] * len(sentence_word_counts(conclusion)),
    "Jumlah Kata": sentence_word_counts(abstract)
                  + sentence_word_counts(introduction)
                  + sentence_word_counts(conclusion)
})

# Tampilkan tabel rapi
print("ðŸ“Š Jumlah Kata per Kalimat di Tiap Bagian:")
print(data)

# Histogram per bagian
plt.figure(figsize=(8,5))
for section in ["Abstract", "Introduction", "Conclusion"]:
    subset = data[data["Bagian"] == section]
    plt.hist(subset["Jumlah Kata"], bins=range(0, max(data["Jumlah Kata"])+5, 5),
             alpha=0.6, label=section)

plt.title("Distribusi Jumlah Kata per Kalimat di Tiap Bagian")
plt.xlabel("Jumlah Kata per Kalimat")
plt.ylabel("Frekuensi")
plt.legend()
plt.show()
```

### Interpretasi Tabel dan Histogram

1. **Abstract**

Pada bagian Abstract, jumlah kata per kalimat cenderung lebih panjang dan bervariasi, bahkan ada satu kalimat yang sangat panjang (sekitar 90 kata). Ini menunjukkan bahwa abstrak berisi kalimat padat dengan informasi kompleks, karena harus menjelaskan tujuan, metode, dan hasil penelitian dalam ruang terbatas.

2. **Introduction**

Introduction punya pola yang lebih seimbang dan terpusat di 15â€“30 kata per kalimat, menandakan gaya penulisan yang informasional tapi mudah dibaca. Penulis tampaknya berusaha menjelaskan konteks penelitian secara detail, namun tetap menjaga keterbacaan.

3. **Conclusion**

Pada Conclusion, panjang kalimatnya lebih pendek dan seragam, biasanya antara 15â€“25 kata. Ini menunjukkan bahwa bagian kesimpulan ditulis dengan to the point, menegaskan hasil utama tanpa perlu uraian panjang.

# 3. Jumlah Kalimat per Paragraf

```python
import matplotlib.pyplot as plt
import re

# Teks yang sudah dipisah
sections = {
    "Abstract": abstract,
    "Introduction": introduction,
    "Conclusion": conclusion
}

sentence_counts = {}

# Hitung jumlah kalimat tiap paragraf
for section, text in sections.items():
    # Pisahkan paragraf berdasarkan newline kosong
    paragraphs = [p.strip() for p in text.split('\n\n') if p.strip()]
    counts = []
    for para in paragraphs:
        # Hitung kalimat berdasarkan titik (.), tanda tanya (?), atau tanda seru (!)
        sentences = re.split(r'[.!?]', para)
        # Hapus elemen kosong
        sentences = [s for s in sentences if s.strip()]
        counts.append(len(sentences))
    sentence_counts[section] = counts

# Cetak hasil per paragraf
for section, counts in sentence_counts.items():
    print(f"\n{section}:")
    for i, count in enumerate(counts, 1):
        print(f"  Paragraf {i}: {count} kalimat")

# Buat histogram
plt.figure(figsize=(12,6))
colors = ['skyblue', 'lightgreen', 'salmon']

for i, (section, counts) in enumerate(sentence_counts.items()):
    plt.bar([x + i*0.25 for x in range(len(counts))], counts, width=0.25, color=colors[i], label=section)

plt.xlabel('Paragraf')
plt.ylabel('Jumlah Kalimat')
plt.title('Jumlah Kalimat per Paragraf di Setiap Bagian')
plt.xticks([x + 0.25 for x in range(max(len(c) for c in sentence_counts.values()))],
           [f"P{x+1}" for x in range(max(len(c) for c in sentence_counts.values()))])
plt.legend()
plt.show()
```

ðŸ“˜ Abstract hanya terdiri dari satu paragraf dengan 6 kalimat, menandakan bagian ini ringkas dan langsung menjelaskan inti penelitian.

ðŸ“— Introduction memiliki 8 paragraf dengan jumlah kalimat bervariasi antara 4 hingga 9 kalimat. Paragraf kedua adalah yang terpanjang (9 kalimat), berisi latar belakang atau landasan teori yang cukup mendalam. Pola ini menunjukkan bahwa bagian pendahuluan memiliki struktur yang cukup padat dan terperinci di awal, lalu menurun di bagian akhir yang lebih singkat.

ðŸ“• Conclusion terdiri dari 4 paragraf, dengan jumlah kalimat antara 4 sampai 8. Paragraf pertama paling panjang (8 kalimat), berisi rangkuman utama hasil penelitian, sedangkan paragraf selanjutnya berfungsi untuk menegaskan kesimpulan dan implikasi secara lebih ringkas.

# Pairplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

data = {'Bagian': ['Abstract', 'Introduction', 'Conclusion'],
        'text': [abstract, introduction, conclusion]}
df = pd.DataFrame(data)

# Fungsi analisis per kalimat
def analyze_sentences(text):
    sentences = re.split(r'[.!?]', text)
    data = []
    for s in sentences:
        s = s.strip()
        if len(s) > 0:
            words = s.split()
            num_chars = len(s.replace(" ", ""))
            num_vowels = len(re.findall(r'[aeiouAEIOU]', s))
            num_consonants = num_chars - num_vowels
            data.append({
                'Kalimat': s,
                'Panjang_Kalimat': len(words),
                'Rata2_Panjang_Kata': num_chars / len(words),
                'Vokal': num_vowels,
                'Konsonan': num_consonants
            })
    return pd.DataFrame(data)

# Pisahkan berdasarkan bagian dan analisis
dfs = []
for index, row in df.iterrows():
    temp = analyze_sentences(row['text'])
    temp['Bagian'] = row['Bagian']
    dfs.append(temp)

final_df = pd.concat(dfs, ignore_index=True)

# Pairplot
sns.pairplot(final_df[['Vokal','Konsonan','Rata2_Panjang_Kata','Panjang_Kalimat','Bagian']],
             hue='Bagian', diag_kind='kde', corner=True)
plt.suptitle("Hubungan antara Vokal, Konsonan, Panjang Kata, dan Panjang Kalimat", y=1.02)
plt.show()
```

### **1. Hubungan antar variabel utama**

* Terlihat adanya **hubungan positif yang kuat antara jumlah vokal, konsonan, dan panjang kalimat**. Artinya, semakin banyak vokal dan konsonan dalam sebuah kalimat, semakin panjang pula kalimat tersebut.
* Pola titik yang membentuk garis diagonal di scatterplot â€œVokal vs Panjang Kalimatâ€ dan â€œKonsonan vs Panjang Kalimatâ€ memperkuat indikasi bahwa kalimat panjang tentu mengandung lebih banyak huruf secara proporsional.

---

### **2. Rata-rata panjang kata (Rata2_Panjang_Kata)**

* Sebagian besar kalimat memiliki **rata-rata panjang kata antara 4â€“6 huruf**, dan distribusinya relatif mirip di ketiga bagian.
* Tidak ada hubungan yang terlalu kuat antara panjang kata dengan panjang kalimat â€” kalimat panjang belum tentu tersusun dari kata-kata panjang, tetapi dari jumlah kata yang lebih banyak.

### **3. Perbedaan antar bagian artikel**

* **Introduction (oranye)** memiliki sebaran data paling luas, menunjukkan variasi kalimat yang beragam â€” dari yang sangat pendek hingga panjang. Ini menggambarkan struktur pengantar yang fleksibel, terdiri dari kalimat ringkas pembuka hingga kalimat penjelasan yang panjang.
